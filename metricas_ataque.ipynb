```python
import networkx as nx
import random
import matplotlib.pyplot as plt
```

```python
er_graph = nx.erdos_renyi_graph(1000, 0.3)
ba_graph = nx.barabasi_albert_graph(1000, 4)
```

```python
def calcula_distancia_media(graph):
    if nx.is_connected(graph):
        return nx.average_shortest_path_length(graph)
    else:
        return float('inf')
```
