```python
import networkx as nx
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
import random
import copy
```

```python
def update(frame, G, pos, ax, states, history):
    ax.clear() 
    ax.set_title(f"Time Step: {frame + 1}")

    node_colors = []
    for node in G.nodes:
        if history[frame][node] == 'S':
            node_colors.append('blue')
        elif history[frame][node] == 'I':
            node_colors.append('red')
        elif history[frame][node] == 'R':
            node_colors.append('green')
    nx.draw(G, pos, with_labels=True, node_color=node_colors, ax=ax)

def simulate_epidemic(G, beta, gamma, timesteps):
    states = {node: 'S' for node in G.nodes}
    initial_infected = random.choice(list(G.nodes))
    states[initial_infected] = 'I'
    
    history = []

    for t in range(timesteps):
        history.append(states.copy())
        new_states = states.copy()
        
        for node in G.nodes:
            if states[node] == 'I':
                for neighbor in G.neighbors(node):
                    if states[neighbor] == 'S' and random.random() < beta:
                        new_states[neighbor] = 'I'
                if random.random() < gamma:
                    new_states[node] = 'R'
        states = new_states
    
    return history

def animate_epidemic(G, beta, gamma, timesteps):
    pos = nx.spring_layout(G)
    history = simulate_epidemic(G, beta, gamma, timesteps)
    
    fig, ax = plt.subplots()
    
    ani = FuncAnimation(fig, update, frames=timesteps, fargs=(G, pos, ax, history[0], history), repeat=False, interval=1000)
    
    return HTML(ani.to_html5_video())

random_graph = nx.erdos_renyi_graph(20, 0.3)

beta = 0.2  
gamma = 0.1 
timesteps = 10 

animate_epidemic(random_graph, beta, gamma, timesteps)
```
